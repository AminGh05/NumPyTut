# Random in NumPy

Sure we're not going to work with real randomness and it is sudo-random we're using. Random functions are absolutely handy in statistics and distributions we're gonna be using for ML.

## Random number in NumPy

Use `random` module from numpy to work with them.
- `randint()` for random integer in the interval given
- `rand()` for random float in [0, 1]

```python
from numpy import random

x = random.randint(100) # generate a random int in [0, 100]
print(x)

y = random.rand() # generate a random float in [0, 1]
print(y)
```

### Random Arrays

It's easy to generate random arrays with `random` module in NumPy. Just add parameter `size` to the method as a tuple and get the result.


```python
x = random.randint(100, size=(5))
print(x)

x = random.randint(100, size=(2, 3)) # 2D array
print(x)

x = random.rand(5)
print(x)

x = random.rand(2, 3)
print(x)
```

## Choice

Chooses an element randomly from the iterable given.

**Note:** It can return an array of random elements from the input, by giving the `size` argument.

```python
x = random.choice([1, 2, 3, 4])
print(x)

y = random.choice([3, 4, 5, 6], size=(3, 3))
print(y)
```

### Add Probablity

Based on the **distribution** and *PMF* or *PDF*, we can give an iterable if probablities to choice function.

**Note:** 0 means no instance of the element

```python
# 1D array of 100 number with the distribution told
arr = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))
print(arr)

# make it 2D
arr = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(10, 10))
print(arr)
```

## Random Permutation

Considering the order of elements matters in arrays, [1, 2, 3] and [3, 2, 1] are both permutations of 1, 2, 3

To generate permutations of elements, use `shuffle` or `permutation` methods

**Note:** `shuffle` does the operation on the array itself while, `permutation` returns a copy

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)
print(arr)

arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
```

## Plotting

Using `matplotlib` and `seaborn`, we're gonna plot our random distributions

## Normal Distribution

One of the most important distributions with many natural examples

Use `random.normal()` with these parameters:
- `loc` which is the mean
- `scale` -> the standard deviation
- `size` -> the shape of returned array

```python
X = random.normal(loc=1, scale=2, size=(100000))
sns.displot(X, kind="kde")
plt.show()
```

## Binomial Distribution

A discrete distribution, an expnansion of Bernoulli dist.

Use `random.binomial()` with these parameters:
- `n` which is the number of trials
- `p` -> probablity of occurence
- `size` -> the shape of returned array

```python
sns.displot(random.binomial(n=10, p=0.5, size=10000))
plt.show()
```

### Normal vs. Binomial

Binomial with enough points to lead to infinity, can be like a normal dist.
It is a consequence of **Central Limit Theorem**

```python
data = {
    "normal": random.normal(loc=50, scale=5, size=1000),
    "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data=data, kind="kde")
plt.show()
```

## Poisson Distribution

A discrete dist to measure how many times an event can happen in a specified time

Use `random.poisson()` with these parameters:
- `lam`: rate
- `size`: the shape of the returned array

```python
sns.displot(random.poisson(lam=4, size=10000))
plt.show()
```

## Uniform Distribution

Every event has equal chance of happening

Use `random.uniform()` with these parameters:
- `low`: lower bound - defuault 0.0
- `high`: upper bound - default 1.0
- `size` -> the shape of returned array

```python
sns.displot(random.uniform(size=10000), kind="kde")
plt.show()
```

## Logistic Distribution

Used to show **growth**. It's so much like the ***normal*** dist, but with heavier **tails**.

Use `random.logistic()` with these parameters:
- `loc` which is the number mean
- `scale` which is the standard deviation 
- `size` -> the shape of returned array

```python
sns.displot(random.logistic(size=10000), kind="kde")
plt.show()
```

```python
# compare normal and logistic
data = {
    "normal": random.normal(scale=2, size=10000),
    "logistic": random.logistic(size=10000)
}

sns.displot(data=data, kind="kde")
plt.show()
```

## Multinomial Distribution

General form of ***Binomial*** with several scenarios.

Use `random.multinomial()` with these parameters:
- `n` -> number of times to run the experiment
- `pvals` -> list of probablities 
- `size` -> the shape of returned array

**Note:** Multinomial samples will produce one value for each pval

```python
x = random.multinomial(n=10000, pvals=[1/6] * 6)
print(x)

sns.displot(x, kind='kde')
plt.show()
```

## Exponential Distribution

Used to show the time till the next event. Similar to Poisson in continuous space.

Use `random.exponential()` with these parameters:
- `scale` which is the inverse of rate - like the **lambda** in poisson. 
- `size` -> the shape of returned array

**Note:** Poisson distribution deals with number of occurences of an event in a time period whereas exponential distribution deals with the time between these events.

```python
sns.displot(random.exponential(scale=2 ,size=10000), kind="kde")
plt.show()
```
